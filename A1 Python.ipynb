```python
import pandas as pd
df = pd.read_excel("APM466 A1.xlsx")
```

```python
from datetime import date
from datetime import datetime
```

```python
ttm = []
for i in range(len(df["maturity date"])):
    ttm.append((df["maturity date"][i].to_pydatetime()-datetime.today()).days//30+1)
df["month_to_maturity"] = ttm
df
```

```python
days = ["1.10","1.11","1.12","1.13","1.14","1.17","1.18","1.19","1.20","1.21","1.24"]
for i in range(len(days)):
    day_since_coupon = []
    for j in range(len(df["yield"])):
        diff = (df.columns[7+i]-df["last coupon date"][j]).days
        day_since_coupon.append(diff)
    df["days_since_last_coupon{}".format(days[i])] = day_since_coupon
```

```python
days = ["1.10","1.11","1.12","1.13","1.14","1.17","1.18","1.19","1.20","1.21","1.24"]
for i in range(len(days)):
    AIs = []
    for j in range(len(df["yield"])):
        AI = df["coupon"][j] * 100 * df.iloc[:,19+i][j]/365
        AIs.append(AI)
    df["AI{}".format(days[i])] = AIs
```

```python
days = ["1.10","1.11","1.12","1.13","1.14","1.17","1.18","1.19","1.20","1.21","1.24"]
for i in range(len(days)):
    Cleans = []
    for j in range(len(df["yield"])):
        Clean = df.iloc[:,7+i][j] - df.iloc[:,30+i][j]
        Cleans.append(Clean)
    df["Clean{}".format(days[i])] = Cleans
df.columns
```

```python
df['month_since_last_coupon'] = df['days_since_last_coupon1.11']//30
```
